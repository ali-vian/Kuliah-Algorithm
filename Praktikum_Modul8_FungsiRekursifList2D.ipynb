### 1.1	Penjumlahan List
Buatlah Fungsi Rekursif untuk menjumlahkan semua anggota yang terdapat pada suatu list, dengan tampilan seperti yang ditunjukkan pada Gambar 1


```python
# Jawaban
def sumList(numList):
    if len(numList)==1:
        return numList[-1]
    else:
        print(numList[:-1],"+",numList[-1])
        return sumList(numList[:-1])+numList[-1]
```

### 1.2	Penjumlahan Karakter Angka di dalam String
Buatlah Fungsi Rekursif untuk menjumlahan setiap digit karakter yang berupa angka di dalam sebuah string. [Hint] Gunakan fungsi string.isdigit() untuk pengecakan apakah karakter di dalam string adalah angka. Tampilan output dari fungsi ini dapat dilihat pada Gambar 2


```python
# Jawaban
def sumStringDigit(strAngka):
    if strAngka.isdigit():
        if len(strAngka)==1:
            return int(strAngka[0])
        else:
            return int(strAngka[0])+int(sumStringDigit(strAngka[1:]))
    else:
        return "Tidak dapat dikonversi menjadi angka"
```

```python
strNum='-42357'
print("Total penjumlahan dari digit '",strNum,"' adalah ",sumStringDigit(strNum))
```

### 2.1	Create List 2D
Buatlah fungsi untuk membuat list 2D, dimana setiap anggotanya merupakan inputan dari user. Buatlah code agar memiliki tampilan seperti pada Gambar 3, dan return value berupa list2D

(a)	Inputan

(b)	Output

(d)	Ouput
Gambar 3: Create List 2D

```python
# jawaban
def createList2D(baris,kolom):
    newList2D=[]
    for i in range(baris):
        newList=[]
        for j in range(kolom):
            newList.append(int(input(f"list [{i},{j}] : ")))
        newList2D.append(newList)
    return newList2D
```
