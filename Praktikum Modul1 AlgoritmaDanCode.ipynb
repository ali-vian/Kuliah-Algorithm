#### 1. Type Data
>Di dalam bahasa python terdapat 4 type data yaitu: 
>- Integer (int) adalah jenis type data yang digunakan untuk menyimpan data dalam bentuk angka (bilangan bulat) dan bukan pecahan 
>- float adalah jenis type data yang digunakan untuk menyimpan data dalam bentuk angka pecahan (decimal)

#### 2. Algoritma 
> Definisi<br>
>Algoritma adalah langkah-langkah dalam menyelesaikan suatu masalah dengan cara membuat input menjadi output yang di inginkan. Algoritma inilah yang akan di ubah menjadi syntax dalam bahasa pemrograman. oleh karena itu, sebelum membuat program kita harus membuat algoritma terlebih dahulu.
    
>Jenis - jenis Algoritma<br>
Terdapat tiga jenis algoritma yaitu :
>- Sequence adalah jenis algoritma yang di jalankan secara berurutan dari awal hingga akhir 
>- Percabangan (braching) pada jenis algoritma ini terdapat pilihan, dalam percabangan ini terdapat kondisi True dan False yang mana jika memenuhi kondisi True maka pilihan yang true yang di kerjakan dan sebaliknya.
>- Perulangan (Looping) pada jenis algoritma ini dikerjakan secara berulang ulang selama kondisi looping bernilai True dan akan berhenti jika bernilai False.

>Bagaimana menulis algoritma<br>
>- Pertama tentukan masalah apa yang akan selesaikan
>- Tentukan inputan yang akan diterima 
>- tentukan dengan cara apa masalah di selesaikan
>- output yang keluar dalam bentuk apa

## 2.1 Program Bilangan Ganjil 
Konversi algoritma yang sudah ditulis sebelumnya dengan menggunakan bahasa Pemrograman Python.
Contoh hasil eksekusi dapat dilihat di Gambar.2a or Gambar.2b :


```python
#Jawaban

n=int(input("Masukkan bilangan ="))
jumlah=0
temp=1
for i in range(1,n*2,2):
    print("Bilangan ke- ",temp," : ",i)
    jumlah+=i
    temp+=1
print("T0tal = ",jumlah)
```

### Program Deret Aritmatika 
Konversi algoritma yang sudah ditulis sebelumnya dengan menggunakan bahasa Pemrograman Python,
seperti yang ditunjukkan pada Gambar 3a or Gambar3b.


```python
#jawaban

n=int(input("masukkan nilai n = "))
a=int(input("masukkan nilai a = "))
b=int(input("masukkan nilai b = "))
sn=temp=0
for i in range(a,n*b+a,b):
    temp+=1
    print("u- ",temp," = ",i)
    sn+=i
print("Sn = ",sn)
```
